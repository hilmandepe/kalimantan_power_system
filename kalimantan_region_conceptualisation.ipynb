```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import folium
import math
import networkx as nx
import import_ipynb
```

```python
initial_network_nodes = pd.read_csv('database/output/buses.csv', sep = ",", quotechar = "'")

initial_network_nodes
```

```python
initial_network = pd.read_csv('database/output/lines.csv', sep = ",", quotechar = "'")

initial_network
```

```python
combined = initial_network.merge(initial_network_nodes, left_on = "bus0", right_on = "node", suffixes = ("_left", "_right"))
combined = combined.merge(initial_network_nodes, left_on  = "bus1", right_on = "node", suffixes = ("_0", "_1"))
combined = combined [["bus0", "lon0", "lat0", "bus1", "lon1", "lat1"]]

display(combined)
```

```python
basemap = folium.Map(location = [-0.152375, 113.899132], zoom_start = 6)

maxvoltage = max(initial_network_nodes.v_nom)

for i in range(len(initial_network_nodes)):
    popupinfo = (initial_network_nodes.node.iloc[i])
    folium.CircleMarker(location = [initial_network_nodes.lat.iloc[i], initial_network_nodes.lon.iloc[i]], 
                       tooltip = popupinfo, color = "green", fill = True, 
                       radius = math.sqrt(10 * initial_network_nodes.v_nom.iloc[i] / maxvoltage)).add_to(basemap)

for i in range(len(combined)):
    points = [[combined.lat0.iloc[i], combined.lon0.iloc[i]],
             [combined.lat1.iloc[i], combined.lon1.iloc[i]]]
    folium.PolyLine(points, color = "yellow", opacitiy = 0.7).add_to(basemap)
    
basemap.save("concept_nodes.html")

display(basemap)
```
